```python
#Checking for missing values
df.isnull().sum()
```

```python
#Removing missing values
df_clean = df.dropna()
```

```python
#Checking if the missing values have been removed
df_clean.isnull().sum()
```

```python
#Chaging data type - Year from float to integer
df_clean['Year'] = df_clean['Year'].astype(int)
```

```python
#Checking the clean Data Frame
df_clean
```

```python
# Rename columns names
df_clean.rename(columns={'Lead Studio':'Studio'})

```

```python
# Your DataFrame with two columns 'column1' and 'column2'
df_clean['Profitability'] = df['Profitability'].apply(lambda x: "${:,.6f}".format(x))
df_clean['Worldwide Gross'] = df['Worldwide Gross'].apply(lambda x: "${:,.6f}".format(x))
```

```python
#Checking the final result of the cleaning and formatting
df_clean
```

```python
#Export the clean data into my pc
from google.colab import files

df_clean.to_csv('Cleaned_Data.csv', index=False)

# Create a download link
files.download('Cleaned_Data.csv')
```
